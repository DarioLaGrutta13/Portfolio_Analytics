```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\dario\\OneDrive\\Escritorio\\Unab\\Inferencia y Patrones\\Comisi√≥n 2 Florencia Statti\\TP1\winequality-red.csv",
                 header=0, sep= ';')
df.head()
```

```python
standard = StandardScaler()
dfStandard = pd.DataFrame(standard.fit_transform(df), columns=df.columns)
dfStandard.head()
```

```python
X= dfStandard.drop("quality", axis=1)
y=df["quality"]
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
quality = y.unique().tolist()
quality.sort()
```

```python
plt.figure(figsize=(12,10))
for i in range(len(quality)):
    plt.scatter(X_pca[y == quality[i], 0], X_pca[y == quality[i], 1], label=quality[i])
plt.xlabel('Primera componente principal')
plt.ylabel('Segunda componente principal')
plt.legend(title='Calidad del vino')
plt.title('PCA del conjunto Wine')
plt.show()
```

```python
for i in range(len(quality)):
    q= df[df["quality"] == quality[i]]
    cant = len(q["quality"].tolist())
    print(f"\ncalidad de vino: {quality[i]} \nCantidad de muestras: {cant}")
```
