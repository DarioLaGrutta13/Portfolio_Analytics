```python
import pandas as pd
import numpy as np
```

```python
spo = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='latin-1')
spo.info()
```

```python
# 1. Reemplazar las comas (',') con nada para que los valores sean numéricos válidos.
spo['Spotify Streams'] = spo['Spotify Streams'].astype(str).str.replace(',', '', regex=True)
spo['Spotify Playlist Count'] = spo['Spotify Playlist Count'].astype(str).str.replace(',', '', regex=True)
spo['Spotify Playlist Reach'] = spo['Spotify Playlist Reach'].astype(str).str.replace(',', '', regex=True)
spo['YouTube Views'] = spo['YouTube Views'].astype(str).str.replace(',', '', regex=True)
spo['YouTube Likes'] = spo['YouTube Likes'].astype(str).str.replace(',', '', regex=True)
spo['TikTok Posts'] = spo['TikTok Posts'].astype(str).str.replace(',', '', regex=True)
spo['TikTok Likes'] = spo['TikTok Likes'].astype(str).str.replace(',', '', regex=True)
spo['TikTok Views'] = spo['TikTok Views'].astype(str).str.replace(',', '', regex=True)
spo['YouTube Playlist Reach'] = spo['YouTube Playlist Reach'].astype(str).str.replace(',', '', regex=True)

spo['Apple Music Playlist Count'] = spo['Apple Music Playlist Count'].astype(str).str.replace(',', '', regex=True)
spo['Spotify Playlist Count'] = spo['Spotify Playlist Count'].astype(str).str.replace(',', '', regex=True)


# 2. Convertir la columna al tipo numérico (float).
#    Esto convierte automáticamente las cadenas 'nan' y los valores faltantes originales
#    (si los había) a np.nan.
spo['Spotify Streams'] = pd.to_numeric(spo['Spotify Streams'], errors='coerce')
spo['Spotify Playlist Count'] = pd.to_numeric(spo['Spotify Playlist Count'], errors='coerce')
spo['Spotify Playlist Reach'] = pd.to_numeric(spo['Spotify Playlist Reach'], errors='coerce')
spo['YouTube Views'] = pd.to_numeric(spo['YouTube Views'], errors='coerce')
spo['YouTube Likes'] = pd.to_numeric(spo['YouTube Likes'], errors='coerce')
spo['TikTok Posts'] = pd.to_numeric(spo['TikTok Posts'], errors='coerce')
spo['TikTok Likes'] = pd.to_numeric(spo['TikTok Likes'], errors='coerce')
spo['TikTok Views'] = pd.to_numeric(spo['TikTok Views'], errors='coerce')
spo['YouTube Playlist Reach']=pd.to_numeric(spo['YouTube Playlist Reach'], errors='coerce')

spo['Apple Music Playlist Count'] = pd.to_numeric(spo['Apple Music Playlist Count'], errors='coerce')
spo['Spotify Playlist Count']=pd.to_numeric(spo['Spotify Playlist Count'], errors='coerce')


# O más simple, usa el método fillna de Pandas (que es más idiomático):
spo['Spotify Streams'] = spo['Spotify Streams'].fillna(0)
spo['Spotify Playlist Count'] = spo['Spotify Playlist Count'].fillna(0)
spo['Spotify Playlist Reach'] = spo['Spotify Playlist Reach'].fillna(0)
spo['YouTube Views'] = spo['YouTube Views'].fillna(0)
spo['YouTube Likes'] = spo['YouTube Likes'].fillna(0)
spo['TikTok Posts'] = spo['TikTok Posts'].fillna(0)
spo['TikTok Likes'] = spo['TikTok Likes'].fillna(0)
spo['TikTok Views'] = spo['TikTok Views'].fillna(0)
spo['YouTube Playlist Reach']=spo['YouTube Playlist Reach'].fillna(0) 

spo['Apple Music Playlist Count'] = spo['Apple Music Playlist Count'].fillna(0)
spo['Spotify Playlist Count'] = spo['Spotify Playlist Count'].fillna(0)

spo.info()
```

```python
spo.to_csv("Most Streamed Spotify Songs 2024 corregido.csv", index=False, encoding='utf-8')
```
