```python
import pandas as pd
import numpy as np
```

```python
spo = pd.read_csv("spotify-2023.csv", encoding='latin-1')
spo.info()
```

```python
# 1. Reemplazar las comas (',') con nada para que los valores sean numéricos válidos.
spo['in_shazam_charts'] = spo['in_shazam_charts'].astype(str).str.replace(',', '', regex=True)
spo['streams'] = spo['streams'].astype(str).str.replace(',', '', regex=True)
spo['in_deezer_playlists'] = spo['in_deezer_playlists'].astype(str).str.replace(',', '', regex=True)

# 2. Convertir la columna al tipo numérico (float).
#    Esto convierte automáticamente las cadenas 'nan' y los valores faltantes originales
#    (si los había) a np.nan.
spo['in_shazam_charts'] = pd.to_numeric(spo['in_shazam_charts'], errors='coerce')
spo['streams'] = pd.to_numeric(spo['streams'], errors='coerce')
spo['in_deezer_playlists'] = pd.to_numeric(spo['in_deezer_playlists'], errors='coerce')


# 3. Ahora que la columna es numérica, puedes crear el array de NumPy (si es necesario)
#    o trabajar directamente con la columna de Pandas.
shazam = spo['in_shazam_charts'].values 

# Ahora, np.where y np.isnan funcionarán:
#in_shazam_charts_corrected = np.where(np.isnan(shazam), 0, shazam)

# O más simple, usa el método fillna de Pandas (que es más idiomático):
spo['in_shazam_charts_final'] = spo['in_shazam_charts'].fillna(0)
spo['streams'] = spo['streams'].fillna(0)
```

```python
# Usar .astype(str) para garantizar que es una cadena estándar de Python/UTF-8
spo['artist(s)_name'] = spo['artist(s)_name'].astype(str)
spo['track_name'] = spo['track_name'].astype(str)
# ... y cualquier otra columna de texto que muestre problemas.
```

```python
spo.to_csv("spotify-2023_corregido.csv", index=False, encoding='utf-8')
```
