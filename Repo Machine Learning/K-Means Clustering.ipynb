```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
#inventamos la base de datos 
np.random.seed(42)
data = np.random.rand(100,2)
data
```

```python
kmeans = KMeans(n_clusters = 5 , random_state = 42)
kmeans.fit(data)
```

```python
#genero los centroides a utilizar
centroides = kmeans.cluster_centers_
```

```python
#genero las etiquetas para identificar a cada valor con un centroide
etiquetas = kmeans.labels_
etiquetas
```

```python
#Generar grafico de dispersion
plt.figure(figsize=(10,6))
plt.scatter(data[:,0], data[:,1], c = etiquetas, cmap="viridis", marker = "o")

#genero la visualizacion de los centroides
plt.scatter(centroides[:,0], centroides[:,1], marker = "x", c = "red", s=200, label = "Centroides")
plt.title("Visualizacion de K-Means Clusteriongs")
plt.legend();
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos.dropna(inplace = True)
```

```python
data = pinguinos[["bill_length_mm", "bill_depth_mm"]]
```

```python
data
```

```python
kmeans = KMeans(n_clusters = 3, random_state= 42)
kmeans.fit(data)

```

```python
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data=pinguinos, x= "bill_length_mm", y ="bill_depth_mm", hue = etiquetas, palette = "viridis")
plt.scatter(centroides[:,0],centroides[:,1],c="red",marker="x", s=200, label="Centroides")
plt.title("Visualizacion de K-Means")
plt.legend()
plt.xlabel("Longitud del pico (mm)")
plt.ylabel("Profundidad del pico (mm)")
```
