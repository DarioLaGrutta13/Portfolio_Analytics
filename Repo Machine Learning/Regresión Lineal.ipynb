```python
import pandas as pd
import matplotlib.pyplot as plt

#sklearn es la biblioteca de machine learning
#en este caso importamos el modelo lineal

from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

La regresion lineal busca conformar una funcion lineal que se acerque lo mas posible a futuros datos (predecir) y describa las relaciones basados en dos variables justamente relacionadas (entrada y salida). 

```python
plt.scatter(df["Area"], df["Precio"]);
```

```python
#entrenamos a nuestro algoritmo para que pueda
#realizar una regresion lineal donde logre trazar
#una linea representativa con el menor margen de 
#error para lograr establecer los valores promedios

X = df[["Area"]]

#Se pone doble [[]] porque las bibliotecas de machine
#learning pueden recibir varias columnas, entonces
#debo dejar el formato como si fuera a pasar justamente
#varias columnas. Ejemplo
# DataFrame[[Col1],[Col2],...,[Coln]]
#para este caso, solo pasamos una columna que es Area
#Se debe poner doble [] para que la variable sea del
#tipo DataFrame, si solo pongo uno, pasaria a ser una
#Serie

#Por convencion se deja la variable en mayuscula para
#identificar que pasaremos un DataFrame y no una Serie
```

```python
y = df["Precio"]

#La variable de salida si tiene que ser una Serie
```

```python
#Creamos un objeto que sea el modelo de regresion

modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y)
#el comando fit realiza el entrenamiento de la regresion
# le paso como parametro 1° las variables de entradas
# y 2° las de salida o dependientes
```

```python
modelo.predict([[3300]])
#aca lo que estamos haciendo es activar la prediccion
#le estamos dando un valor que vendrian a ser los
#m2 del area, y nos devuelve el precio al que estario
#un inmueble aproximadamente
```

```python
plt.plot(df["Area"],
        modelo.predict(X));

#Graficamos la linea de regresion pasando como 
#variable independiente el Area, pero como resultados
#pedimos que se tomen las predicciones generadas
```

```python
#para ver que tan acertadas fueron las predicciones
#graficamos ambas situaciones a la vez

plt.scatter(df["Area"], df["Precio"])

plt.plot(df["Area"],
        modelo.predict(X));
```
