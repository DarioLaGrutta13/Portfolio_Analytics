# Arboles de decision

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
#prueba con el data set iris, que presenta registro de flores

df = sns.load_dataset("iris")
df.head()
```

```python
df.species.unique()
```

Se creará un programa que indique luego de ingresar las caracteristicas de una nueva flor, si sera de la categoria setosa, versicolor o virginica

```python
#definimos las variables dependientes e independientes
X = df.drop("species", axis = 1)
X.head()
```

```python
#para poder establecer un parámetro de referencia para el modelo, transformamos las especies en números claves

le = LabelEncoder()
y = le.fit_transform(df["species"])
y
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X,
                                                            y, 
                                                            train_size=0.8,
                                                           random_state = 42)
```

```python
arbol = DecisionTreeClassifier()
arbol.fit(X_entrena, y_entrena)
```

```python
#from sklearn.tree import  plot_tree es importado 
#para poder graficar el arbol hecho
plt.figure(figsize = (20,20)) #aca agrandamos el grafico
plot_tree(decision_tree = arbol)
plt.show()
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size = 0.8, random_state = 13)


# Creación y entrenamiento del modelo de árbol de decisión
arbol = DecisionTreeClassifier()
arbol.fit(X_entrena, y_entrena)

plt.figure(figsize=(20, 20))

plot_tree(decision_tree = arbol);
```

```python
#from sklearn.tree import  plot_tree es importado 
#para poder graficar el arbol hecho
plt.figure(figsize = (20,20)) #aca agrandamos el grafico
plot_tree(decision_tree = arbol,
         filled = True); 
#con filled podemos darle color a las especies
```

```python
plt.figure(figsize = (20,20)) #aca agrandamos el grafico
plot_tree(decision_tree = arbol,
         filled = True,
         class_names = ["setosa", 
                        "versicolor", 
                        "virgincia"]);
#class_name agrega etiquetas para saber que
#nombre es para cada color
```
