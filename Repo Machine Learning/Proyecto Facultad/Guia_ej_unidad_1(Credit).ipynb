```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
ruta = "C:/Users/dario/OneDrive/Escritorio/Unab/Inferencia y Patrones/Datos para Ejercicios/credit.txt"
df = pd.read_csv(ruta, sep=",", header = 0)
df.info
#nos dice que el df tiene [400 rows x 12 columns]>
```

```python
X = df[["Income","Limit","Rating","Cards","Age","Education"]]
X.head()
```

```python
y = df[["Balance"]]
y.head()
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
X = sm.add_constant(X)
modelo = sm.OLS(y,X).fit()
```

```python
print(modelo.summary())
```

Los p-valores nos dicen que solo es signicativo utilizar el ingreso (income)

```python
X = df[["Income"]]
X = sm.add_constant(X)
modelo = sm.OLS(y,X).fit()
print(modelo.summary())
```

```python
#Supongamos que la varianza de los errores εi del modelo es σ**2. Estimarla.

#Buscamos los residuos
residuos = sum((modelo.resid)**2)
residuos
```

```python
#Cuento los n y los p
n = 400 #Filas observadas
p = 2
```

```python
X #Lo llamamos para observar los p analizados
```
