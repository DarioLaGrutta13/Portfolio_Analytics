```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
iris
```

```python
#el comando .data, me trae solo los componentes independientes (variables)
x = iris.data

#el comando .target, me trae solo los componentes dependientes (variables categoricas pero ya convertidas a numeros)
#ejemplo categoria 1 no pasa a verse como "Flor x" sino que se cataloga con un numero 0,1,2...
y = iris.target
```

```python
#aca estandarizamos los valores del conjunto. Lo que hacemos es restarle a todos los valores el promedio de su categoria
#de esa manera todos los datos ocomienzan a medirse desde el mismo punto de partida y si hay diferencias se notan aun mas
x_centrado = x- np.mean(x, axis=0)

#establecemos que nuestro modelo de PCA tendra 2 componentes (eje x e y)
pca = PCA(n_components=2)
#entrenamos nuestro modelo pasandole los datos estandarizados
x_pca = pca.fit_transform(x_centrado)

#dejamos como ayuda de clasificacion a las categorias que trabaja nuestro data set
especies = ["setosa","versicolor","virginica"]
```

```python
plt.figure(figsize=(8,6))
#vamos a generar un for el cual ira ciclando por un rango igual a la cantidad de categorias que haya en el data set. (0,1,2)
for i in range(0,3):
    plt.scatter(x_pca[y == i, 0], x_pca[y == i, 1], label = especies[i])
plt.xlabel("Primer componente principal")    
plt.ylabel("Segunda componente principal") 
plt.legend()
plt.title("PCA de las iris");
```
